```python
%tensorflow_version 2.x
import tensorflow as tf
device_name = tf.test.gpu_device_name()
if device_name != '/device:GPU:0':
  raise SystemError('GPU device not found')
print('Found GPU at: {}'.format(device_name))
```

```python
with tf.device('/device:GPU:0'):
  xy_train = np.loadtxt('/content/sample_data/mnist_train_small.csv', delimiter=',', dtype=np.float32)
  x_train = xy_train[:, 1:]
  y_train = xy_train[:, 0]

  print(x_train, "\nx_train shape:", x_train.shape)
  print(y_train, "\ny_train shape:", y_train.shape)

  nb_classes = 10 # 0 ~ 9
  y_train_one_hot = tf.keras.utils.to_categorical(y_train, nb_classes)
  print("one_hot:", y_train_one_hot)


  tf.model = tf.keras.Sequential()
  tf.model.add(tf.keras.layers.Dense(units=nb_classes, input_dim=784, activation="softmax"))
  tf.model.compile(loss='categorical_crossentropy', optimizer=tf.keras.optimizers.SGD(lr=0.1), metrics=['accuracy'])
  tf.model.summary()

  history = tf.model.fit(x_train, y_train_one_hot, epochs=100)

  xy_test = np.loadtxt('/content/sample_data/mnist_test.csv', delimiter=',', dtype=np.float32)
  x_test = xy_test[:, 1:]
  y_test = xy_test[:, 0]

  pred = tf.keras.backend.eval(tf.argmax(tf.model.predict(x_test), axis=1))
  for p, y in zip(pred, y_test): # this case don't need flatten()
      print("[{}] Prediction: {} True Y: {}".format(p == int(y), p, int(y)))
```
